```
using System;
using System.IO;

var datafile = File.ReadAllText("inputs/input_day2.txt");
```

```
// Test data

var datafile = 
@"
forward 5
down 5
forward 8
up 3
down 8
forward 2
";
```

```
using System.Text.RegularExpressions;

var reader = new StringReader(datafile);

int horizontal = 0;
int depth = 0;

var matcher = new Regex("^(\\w+)\\s{1}(\\d+)$", RegexOptions.Compiled);

string line = null;

while((line = reader.ReadLine()) != null)
{
    var match = matcher.Match(line);

    if(!match.Success)
    {
        continue;
    }

    var command = match.Groups[1].Value;
    var amount = int.Parse(match.Groups[2].Value);

    if(command == "forward")
    {
        horizontal += amount;
        continue;
    }

    if(command == "up")
    {
        depth -= amount;
        continue;
    }

    if(command == "down")
    {
        depth += amount;
        continue;
    }
}

Console.WriteLine(horizontal * depth);
```
