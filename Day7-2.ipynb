```
using System;
using System.Collections.Generic;
using System.IO;
using System.Linq;

var data = File.ReadAllText("inputs/input_day7.txt");
```

```
// Test data
var data = "16,1,2,0,4,2,7,1,2,14";
```

```
var arithmeticProgressionSum = (int number) => number * (1 + number) / 2;
```

```
var horizontalPositions = data.Split(',')
    .Where(_ => !string.IsNullOrWhiteSpace(_))
    .Select(int.Parse)
    .ToList();

var positions = Enumerable.Range(0, horizontalPositions.Max() + 1);

var cheapest = 
    positions
        .Select(
            position => horizontalPositions
                .Select(horizontalPosition => arithmeticProgressionSum(Math.Abs(horizontalPosition - position)))
                .Sum()
        )
        .Min();

Console.WriteLine(cheapest);
```
